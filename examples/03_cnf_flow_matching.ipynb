# CNF Flow Matching

Train a continuous normalizing flow with conditional flow matching (CFM).

```python
import torch
from boltzmann_generators.flows import CNFFlowModel, VelocityField

torch.manual_seed(2)
device = "cpu"

model = CNFFlowModel(VelocityField(dim=2, hidden_dim=32, num_hidden=2, num_freqs=4)).to(device)
x_data = torch.randn(256, 2, device=device)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
for epoch in range(20):
    loss = model.cfm_loss(x_data)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch % 5 == 0:
        print(f"epoch {epoch}: cfm_loss={loss.item():.4f}")

samples, log_q = model.sample(64, device=device, n_steps=16, n_hutchinson=1)
print("Sample shape:", samples.shape, "log_q finite:", torch.isfinite(log_q).all().item())
```
