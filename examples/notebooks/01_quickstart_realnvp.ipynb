# Quickstart: RealNVP + Trainer

Minimal end-to-end Boltzmann Generator training on the 2D double-well potential.

```python
import torch
from boltzmann_generators.energies import DoubleWell2D
from boltzmann_generators.flows import FlowModel, GaussianPrior, RealNVP
from boltzmann_generators.training import TrainConfig, Trainer

torch.manual_seed(0)
device = "cpu"

energy = DoubleWell2D(a=4.0)
model = FlowModel(
    GaussianPrior(2),
    RealNVP(dim=2, num_layers=4, hidden_dim=32, mask="halves"),
).to(device)

config = TrainConfig(n_epochs=30, batch_size=64, lr=1e-3, w_ml=0.0, w_kl=1.0)
trainer = Trainer(model, energy, config, device=device)
history = trainer.fit(x_data=None)
print("Final loss:", history[-1]["loss"])
```
